```python
import pandas as pd
```

```python
file_path_1 = '/content/drive/MyDrive/DATASET/(1_positif)eng_MH_tweet.csv'
df = pd.read_csv(file_path_1)
display(df)
```

```python
display(df.info())
```

```python
drop_kolom = ['conversation_id_str', 'created_at', 'favorite_count', 'id_str', 'image_url', 'in_reply_to_screen_name', 'lang', 'location', 'quote_count', 'reply_count', 'retweet_count', 'tweet_url', 'user_id_str','username']
df_clean = df.drop(drop_kolom, axis=1)
df_clean.head(25)
```

```python
df['full_text'] = df['full_text'].str.lower()
```

```python
import re

def clean_tweet(text):
    text = re.sub(r"http\S+|www.\S+", "", text)              # URL
    text = re.sub(r"@\w+", "", text)                         # Mention
    text = re.sub(r"#\w+", "", text)                         # Hashtag
    text = re.sub(r"\brt\b", "", text)                       # Retweet
    text = re.sub(r"[^\w\s]", "", text)                      # Punctuation
    text = re.sub(r"\d+", "", text)                          # Numbers
    text = re.sub(r"\s+", " ", text).strip()                 # Extra spaces
    return text

df['clean_text'] = df['full_text'].apply(clean_tweet)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

df['clean_text'] = df['clean_text'].apply(
    lambda x: " ".join([word for word in x.split() if word not in stop_words])
)
```

```python
display(df['clean_text'])
```

```python
df['clean_text'].to_csv('(1)_Positif_final_MH.csv', index=False)
```
